```python
# Task 3: Data Visualization
#This notebook demonstrates data visualization techniques using the Iris dataset. We will create various plots to explore the dataset, identify patterns, and gain insights.
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target

# Display the first few rows
data.head()
```

```python
# Visualization 1: Histogram of Sepal Length
#We will create a histogram to visualize the distribution of sepal lengths in the Iris dataset.
```

```python
# Set the style of seaborn
sns.set(style='whitegrid')

# Create a histogram of sepal length
plt.figure(figsize=(10, 6))
sns.histplot(data['sepal length (cm)'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
# Insights
#The histogram shows that sepal lengths are generally distributed around the mean, with a peak around 5 cm. The distribution appears to be slightly right-skewed, indicating that there are fewer instances of larger sepal lengths.
```

```python
# Visualization 2: Box Plot of Sepal Width by Species
#Next, we will use a box plot to compare the distribution of sepal widths for each species of Iris.
```

```python
# Create a box plot of sepal width by species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal width (cm)', data=data, palette='Set2')
plt.title('Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
# Insights
#The box plot indicates that the sepal width varies across species. The Iris Setosa species has a notably wider sepal width compared to Iris Versicolor and Iris Virginica, which are more similar. The plot also shows that the sepal width for Setosa is less variable, as indicated by the shorter whiskers and smaller interquartile range.
```

```python
# Visualization 3: Pair Plot
#We will use a pair plot to visualize the relationships between all numerical features in the Iris dataset, colored by species.
```

```python
# Create a pair plot
plt.figure(figsize=(12, 10))
sns.pairplot(data, hue='species', palette='husl')
plt.suptitle('Pair Plot of Iris Dataset Features', y=1.02)

# Show the plot
plt.show()
```

```python
## Insights
# The pair plot provides an excellent overview of how different features relate to one another across species. 
# We can see that Iris Setosa is well-separated from the other two species in terms of both sepal and petal measurements.
# There is some overlap between Iris Versicolor and Iris Virginica, particularly in the petal measurements, indicating that they may be more challenging to distinguish based solely on these features.
```

```python
# Summary

#In this notebook, I explored the Iris dataset using various visualizations:
#1. **Histogram**: Showed the distribution of sepal lengths, revealing a slight right skew.
#2. **Box Plot**: Compared sepal widths across species, indicating distinct differences among them.
#3. **Pair Plot**: Illustrated relationships between features, highlighting the separation of Iris Setosa from the other species and indicating overlap between Versicolor and Virginica.

#These visualizations provide valuable insights for understanding the characteristics of the Iris dataset and guide further analysis or modeling.
```
